```python
input_dir = '/kaggle/input/phishing-website-detector/phishing.csv'

df_train = pd.read_csv(input_dir)
df_train
```

```python
X = df_train.drop(['class', 'Index'], axis = 1)
y = df_train['class']

X
```

```python
print(f'X : {X.shape}')
print(f'y : {y.shape}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
score_rf=cross_val_score(RandomForestClassifier(n_estimators=10),X, y,cv=10)
print(score_rf)
```

```python
print(f'X_train : {X_train.shape}')
print(f'y_train : {y_train.shape}')
print(f'X_test : {X_test.shape}')
print(f'y_test : {y_test.shape}')
```

```python
rf_model = RandomForestClassifier(
bootstrap = False,
max_depth = 16,
max_features = 'sqrt',
min_samples_leaf = 1,
min_samples_split = 2,
n_estimators = 500)
```

```python
rf_model.fit(X_train, y_train)
```

```python
rf_Grid.best_params_
```

```python
print(f'Train Accuracy - : {rf_model.score(X_train,y_train): 3f}')
print(f'Test Accuracy - : {rf_model.score(X_test,y_test): 3f}')
```
